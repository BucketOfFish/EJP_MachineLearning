We can add titles and labels, plot things on top of each other, and add a legend.

```python
x = np.arange(50)
y = [i**2 for i in x]
y_rand = [i+(np.random.rand()-0.5)*500 for i in y]
y_wrong = [i**1.8 for i in x]
```

```python
plt.scatter(x, y_rand)
plt.plot(x, y, label='correct fit')
plt.plot(x, y_wrong, label='wrong fit')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y_rand)
plt.plot(x, y, label='correct fit')
plt.plot(x, y_wrong, label='wrong fit')
plt.legend()
plt.title("Two curve fits")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim([-500, 4000])
plt.show()
```
