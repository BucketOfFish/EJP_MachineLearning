# Loss Functions

We have some data stored in "Files/data.h5". There are 1000 values for x and 1000 values for y.

```python
import h5py as h5
import numpy as np
```

```python
data = h5.File("Files/data.h5")
```

```python
list(data.keys())
```

```python
x = data['x'][:]
y = data['y'][:]
```

```python
y
```

We can plot these values. You'll learn how to do this later.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x, y)
```

## Creating a Model

Now we need a model that guesses values of y when given x. Let's guess that y = 2*x^2.

```python
def model(x):
    return 2*x**2
```

```python
y_guess = [model(x_i) for x_i in x]
```

```python
plt.scatter(x, y)
plt.scatter(x, y_guess, color='red')
```

This model isn't perfect, but it looks like a good start. Let's define a loss function and calculate the loss.

## Loss Function

We define a loss function here equal to |y - y_guess|, the absolute value of the difference. This loss function is not common, but we'll use it as an example.

```python
def loss(y_guess_vector, y_truth_vector):
    y_diff = [abs(i-j) for i, j in zip(y_guess_vector, y_truth_vector)]
    return sum(y_diff)
```

```python
print(loss(y_guess, y))
```

Is this loss value high or low? It really doesn't matter. All that matters is that the loss value gets lower when we make a more accurate model. Let's try a different model here.

```python
def model2(x):
    return 2*x**1.8 + x*45
```

```python
y_guess2 = [model2(x_i) for x_i in x]
plt.scatter(x, y)
plt.scatter(x, y_guess2, color='red')
```

This model looks a bit better than the one we had before. Let's see if the loss function is lower.

```python
print(loss(y_guess2, y))
```

As you can see, the loss function is lower, meaning that this model fits the data better.
